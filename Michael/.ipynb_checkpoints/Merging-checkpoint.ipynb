```python
import pandas as pd
```

```python
game_data = pd.read_csv('game_data.csv')
game_stats = pd.read_csv('game_stats.csv')
```

```python
cols = [0]
game_stats.drop(game_stats.columns[cols],axis=1,inplace=True)
game_stats
```

```python
cols = [0]
game_data.drop(game_data.columns[cols],axis=1,inplace=True)
game_data
```

```python
win_list = []

for r in game_data.iterrows():
    if r[1]['score_home'] == r[1]['score_away']:
        win_list.append('blank')
    elif r[1]['score_home'] > r[1]['score_away']:
        win_list.append(r[1]['team_home'])
    else:
        win_list.append(r[1]['team_away'])
        
game_data['Winner/tie'] = win_list
game_data = game_data.rename(columns={"schedule_season": "Year", "schedule_week": "Week", "Team": "Favorite"})
game_data
    
```

```python
new_df = pd.merge(game_stats, game_data[['Year','Week','Winner/tie','Favorite','spread_favorite','team_home','team_away']],  how='left', on=['Year','Week','Winner/tie'])
new_df
```

```python
power_rank = pd.read_csv('power_ranks.csv')
```

```python
power_rank
```

```python
cols = [0]
power_rank.drop(power_rank.columns[cols],axis=1,inplace=True)
power_rank
```

```python
w_power_rank = power_rank.rename(columns={"Team": "Winner/tie","Rank": "w_pwr_rnk"})
w_power_rank
```

```python
l_power_rank = power_rank.rename(columns={"Team": "Loser/tie","Rank": "l_pwr_rnk"})
l_power_rank
```

```python
newer_df = pd.merge(new_df, w_power_rank[['Year','Week','Winner/tie','w_pwr_rnk']],  how='left', on=['Year','Week','Winner/tie'])
```

```python
newest_df = pd.merge(newer_df, l_power_rank[['Year','Week','Loser/tie','l_pwr_rnk']],  how='left', on=['Year','Week','Loser/tie'])
```

```python
newest_df = newest_df.rename(columns={'Winner/tie':'Winner', 'Loser/tie':'Loser'})
newest_df
```

```python
df2 = newest_df
df3 = df2.head(6)
#print(df3)
#print(type(df3.iloc[0]["spread_favorite"]))

home = []
point_dif = []
yards_dif = []
turn_dif = []
rank_dif = []
spread = []
win = []
winspread = []

for i in range(0, len(df3)):
    #print(df3.iloc[i]["Winner"])
    dif = df3.iloc[i]["PtsW"] - df3.iloc[i]["PtsL"]
    
    #home team wins
    if(df3.iloc[i]["Winner"] == df3.iloc[i]["team_home"]):
        
        #home team won
        home.append(1)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(1)
        point_dif.append(df3.iloc[i]['net_ptsw'] - df3.iloc[i]['net_ptsl'])
        yards_dif.append(df3.iloc[i]['net_ydsw'] - df3.iloc[i]['net_ydsl'])
        turn_dif.append(df3.iloc[i]['net_tovw'] - df3.iloc[i]['net_tovl'])
        rank_dif.append(df3.iloc[i]['w_pwr_rnk'] - df3.iloc[i]['l_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(df3.iloc[i]['spread_favorite'])
            if(dif > abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(1)
        else:
            spread.append(-int(df3.iloc[i]['spread_favorite']))
            winspread.append(1)
        
        #away data lost
        home.append(0)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(0)
        point_dif.append(-df3.iloc[i]['net_ptsw'] + df3.iloc[i]['net_ptsl'])
        yards_dif.append(-df3.iloc[i]['net_ydsw'] + df3.iloc[i]['net_ydsl'])
        turn_dif.append(-df3.iloc[i]['net_tovw'] + df3.iloc[i]['net_tovl'])
        rank_dif.append(-df3.iloc[i]['w_pwr_rnk'] + df3.iloc[i]['l_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(-df3.iloc[i]['spread_favorite'])
            if(dif < abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(0)
        else:
            spread.append(int(df3.iloc[i]['spread_favorite']))
            winspread.append(0)
        
    #home team loses
    else:
        
        #home team lost
        home.append(1)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(0)
        point_dif.append(-df3.iloc[i]['net_ptsw'] + df3.iloc[i]['net_ptsl'])
        yards_dif.append(-df3.iloc[i]['net_ydsw'] + df3.iloc[i]['net_ydsl'])
        turn_dif.append(-df3.iloc[i]['net_tovw'] + df3.iloc[i]['net_tovl'])
        rank_dif.append(-df3.iloc[i]['w_pwr_rnk'] + df3.iloc[i]['l_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(-df3.iloc[i]['spread_favorite'])
            if(dif < abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(0)
        else:
            spread.append(int(df3.iloc[i]['spread_favorite']))
            winspread.append(0)
            
        #away team won
        home.append(0)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(1)
        point_dif.append(df3.iloc[i]['net_ptsw'] - df3.iloc[i]['net_ptsl'])
        yards_dif.append(df3.iloc[i]['net_ydsw'] - df3.iloc[i]['net_ydsl'])
        turn_dif.append(df3.iloc[i]['net_tovw'] - df3.iloc[i]['net_tovl'])
        rank_dif.append(df3.iloc[i]['w_pwr_rnk'] - df3.iloc[i]['l_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(df3.iloc[i]['spread_favorite'])
            if(dif > abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(1)
        else:
            spread.append(-int(df3.iloc[i]['spread_favorite']))
            winspread.append(1)
            
print(home)
print(point_dif)
print(yards_dif)
print(turn_dif)
print(rank_dif)
print(spread)
print(win)
print(winspread)


#fuck this shit I'm out
```

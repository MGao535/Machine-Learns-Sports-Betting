```python
import pandas as pd
```

```python
df = pd.read_csv("final_df.csv")
df
```

```python
cols = [0]
df.drop(df.columns[cols],axis=1,inplace=True)
df
```

```python
#cols = [0, 3, 4, 5]
#df.drop(df.columns[cols],axis=1,inplace=True)
#df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop(columns=['Win', 'WinSpread'],axis=1)
y = df['Win']
#y = df['WinSpread']
y = y.astype('int')

#Split to test and train 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
score = logreg.score(X_train, y_train)
score2 = logreg.score(X_test, y_test)

print ("Training set accuracy:", '%.3f'%(score))
print ("Test set accuracy:", '%.3f'%(score2))
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow import keras

deep_model = Sequential()
deep_model.add(Dense(units=10, activation='relu', input_dim=6))
deep_model.add(Dense(units=8, activation='relu'))
deep_model.add(Dense(units=6, activation='relu'))
deep_model.add(Dense(units=4, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
opt = keras.optimizers.Adam(learning_rate=0.0005)
```

```python
deep_model.compile(optimizer=opt,
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])
```

```python
deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=200,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = deep_model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

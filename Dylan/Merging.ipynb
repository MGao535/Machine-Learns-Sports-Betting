```python
import pandas as pd
```

```python
game_data = pd.read_csv('game_data.csv')
game_stats = pd.read_csv('game_stats.csv')
```

```python
cols = [0]
game_stats.drop(game_stats.columns[cols],axis=1,inplace=True)
game_stats
```

```python
cols = [0]
game_data.drop(game_data.columns[cols],axis=1,inplace=True)
game_data
```

```python
win_list = []

for r in game_data.iterrows():
    if r[1]['score_home'] == r[1]['score_away']:
        win_list.append('blank')
    elif r[1]['score_home'] > r[1]['score_away']:
        win_list.append(r[1]['team_home'])
    else:
        win_list.append(r[1]['team_away'])
        
game_data['Winner/tie'] = win_list
game_data = game_data.rename(columns={"schedule_season": "Year", "schedule_week": "Week", "Team": "Favorite"})
game_data
    
```

```python
new_df = pd.merge(game_stats, game_data[['Year','Week','Winner/tie','Favorite','spread_favorite','team_home','team_away']],  how='left', on=['Year','Week','Winner/tie'])
new_df
```

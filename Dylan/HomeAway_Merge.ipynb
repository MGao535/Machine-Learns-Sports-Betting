```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned.csv")
df
```

```python
df3 = df

home = []
point_dif = []
yards_dif = []
turn_dif = []
rank_dif = []
spread = []
win = []
winspread = []

for i in range(0, len(df3)):
    #print(df3.iloc[i]["Winner"])
    dif = df3.iloc[i]["PtsW"] - df3.iloc[i]["PtsL"]
    
    #home team wins
    if(df3.iloc[i]["Winner"] == df3.iloc[i]["team_home"]):
        
        #home team won
        home.append(1)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(1)
        point_dif.append(df3.iloc[i]['net_ptsw'])
        yards_dif.append(df3.iloc[i]['net_ydsw'])
        turn_dif.append(df3.iloc[i]['net_tovw'])
        rank_dif.append(df3.iloc[i]['w_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(df3.iloc[i]['spread_favorite'])
            if(dif > abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(1)
        else:
            spread.append(-int(df3.iloc[i]['spread_favorite']))
            winspread.append(1)
        
        #away data lost
        home.append(0)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(0)
        point_dif.append(df3.iloc[i]['net_ptsl'])
        yards_dif.append(df3.iloc[i]['net_ydsl'])
        turn_dif.append(df3.iloc[i]['net_tovl'])
        rank_dif.append(df3.iloc[i]['l_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(-df3.iloc[i]['spread_favorite'])
            if(dif < abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(0)
        else:
            spread.append(int(df3.iloc[i]['spread_favorite']))
            winspread.append(0)
        
    #home team loses
    else:
        
        #home team lost
        home.append(1)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(0)
        point_dif.append(df3.iloc[i]['net_ptsl'])
        yards_dif.append(df3.iloc[i]['net_ydsl'])
        turn_dif.append(df3.iloc[i]['net_tovl'])
        rank_dif.append(df3.iloc[i]['l_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(-df3.iloc[i]['spread_favorite'])
            if(dif < abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(0)
        else:
            spread.append(int(df3.iloc[i]['spread_favorite']))
            winspread.append(0)
            
        #away team won
        home.append(0)
        if(dif == 0):
            win.append(0.5)
        else:
            win.append(1)
        point_dif.append(df3.iloc[i]['net_ptsw'])
        yards_dif.append(df3.iloc[i]['net_ydsw'])
        turn_dif.append(df3.iloc[i]['net_tovw'])
        rank_dif.append(df3.iloc[i]['w_pwr_rnk'])
        if df3.iloc[i]['Winner'] == df3.iloc[i]['Favorite']:
            spread.append(df3.iloc[i]['spread_favorite'])
            if(dif > abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(1)
            elif(dif == abs(df3.iloc[i]['spread_favorite'])):
                winspread.append(0.5)
            else:
                winspread.append(0)
        elif df3.iloc[i]['spread_favorite'] == 0:
            spread.append(0)
            if(dif == 0):
                winspread.append(0.5)
            else:
                winspread.append(1)
        else:
            spread.append(-int(df3.iloc[i]['spread_favorite']))
            winspread.append(1)
            
```

```python
final_df = pd.DataFrame({"home": home, "net_points": point_dif, "net_yard": yards_dif, "net_turn": turn_dif, "weekly_rank": rank_dif, "spread": spread, "Win": win, "WinSpread": winspread})
final_df
```

```python
final_df.to_csv('new_final_df.csv')
```

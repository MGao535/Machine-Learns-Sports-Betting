```python
import pandas as pd
```

```python
game_data = pd.read_csv('game_data.csv')
game_stats = pd.read_csv('game_stats.csv')
```

```python
cols = [0]
game_stats.drop(game_stats.columns[cols],axis=1,inplace=True)
game_stats
```

```python
cols = [0]
game_data.drop(game_data.columns[cols],axis=1,inplace=True)
game_data
```

```python
win_list = []

for r in game_data.iterrows():
    if r[1]['score_home'] == r[1]['score_away']:
        win_list.append('blank')
    elif r[1]['score_home'] > r[1]['score_away']:
        win_list.append(r[1]['team_home'])
    else:
        win_list.append(r[1]['team_away'])
        
game_data['Winner/tie'] = win_list
game_data = game_data.rename(columns={"schedule_season": "Year", "schedule_week": "Week", "Team": "Favorite"})
game_data
    
```

```python
new_df = pd.merge(game_stats, game_data[['Year','Week','Winner/tie','Favorite','spread_favorite','team_home','team_away']],  how='left', on=['Year','Week','Winner/tie'])
new_df
```

```python
power_rank = pd.read_csv('power_ranks.csv')
```

```python
power_rank
```

```python
cols = [0]
power_rank.drop(power_rank.columns[cols],axis=1,inplace=True)
power_rank
```

```python
w_power_rank = power_rank.rename(columns={"Team": "Winner/tie","Rank": "w_pwr_rnk"})
w_power_rank
```

```python
l_power_rank = power_rank.rename(columns={"Team": "Loser/tie","Rank": "l_pwr_rnk"})
l_power_rank
```

```python
newer_df = pd.merge(new_df, w_power_rank[['Year','Week','Winner/tie','w_pwr_rnk']],  how='left', on=['Year','Week','Winner/tie'])
```

```python
newest_df = pd.merge(newer_df, l_power_rank[['Year','Week','Loser/tie','l_pwr_rnk']],  how='left', on=['Year','Week','Loser/tie'])
```

```python
newest_df
```

```python
test_df = newest_df.head(200)
```

```python
test_df.drop(['Week','Date','Loser/tie','Year'],axis=1,inplace=True)
test_df
```

```python
test_df = test_df.rename(columns={'Winner/tie':'Winner'})
```

```python
test_df
```

```python
test_df.loc[test_df.Winner == test_df.team_home,'Winner'] = 2
test_df.loc[test_df.Winner == test_df.team_away,'Winner'] = 1
test_df.head()
```

```python
test_df.loc[test_df.Favorite == test_df.team_home,'Favorite'] = 2
test_df.loc[test_df.Favorite == test_df.team_away,'Favorite'] = 1
test_df.head()
```

```python
test_final = pd.get_dummies(test_df,prefix=['home_team', 'away_team'],columns=['team_home','team_away'])
test_final.head()
```

```python
test_final
```

```python
test_final.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = test_final.drop(columns=['Winner'],axis=1)
y = test_final['Winner']
y = y.astype('int')

#Split to test and train 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
from sklearn.impute import SimpleImputer 
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_train
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
score = logreg.score(X_train, y_train)
score2 = logreg.score(X_test, y_test)

print ("Training set accuracy:", '%.3f'%(score))
print ("Test set accuracy:", '%.3f'%(score2))
```

```python
predictions = logreg.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
real_winner = y_test.to_list()
```
